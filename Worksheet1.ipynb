```python
#Exercise-2
#Task1
import matplotlib.pyplot as plt
image_grayscale = Image . open ("/content/drive/MyDrive/AI and ML/download.jpg")
# Display the grayscale image using matplotlib
plt.imshow(image_grayscale, cmap='gray')
plt.title('Grayscale Image')
plt.show()
```

```python
#Task2
# Convert the Pillow image to a NumPy array, ensuring it's truly grayscale (mode 'L')
image_array_grayscale = np.array(image_grayscale.convert('L'))

# Get the dimensions of the image
height, width = image_array_grayscale.shape

# Calculate the cropping coordinates for a 150x150 section from the center
start_row = height // 2 - 75
end_row = height // 2 + 75
start_col = width // 2 - 75
end_col = width // 2 + 75

# Ensure the coordinates are within image bounds
start_row = max(0, start_row)
end_row = min(height, end_row)
start_col = max(0, start_col)
end_col = min(width, end_col)

# Extract the central 150x150 pixel section using NumPy array slicing
center_section = image_array_grayscale[start_row:end_row, start_col:end_col]

# Display the cropped image using matplotlib
plt.imshow(center_section, cmap='gray')
plt.title('Central 150x150 Pixel Section (Grayscale)')
plt.show()
```

```python
#Task3
# Define the threshold value
threshold_value = 100

# Create a copy to avoid modifying the original grayscale array directly
binary_image_array = image_array_grayscale.copy()

# Apply the threshold: values below 100 to 0, values above 100 to 255
binary_image_array[binary_image_array < threshold_value] = 0
binary_image_array[binary_image_array >= threshold_value] = 255

# Convert the NumPy array back to a PIL Image
binary_image = Image.fromarray(binary_image_array.astype(np.uint8))

# Display the resulting binary image
plt.imshow(binary_image, cmap='gray')
plt.title('Thresholded Binary Image')
plt.show()
```

```python
#Exercise-3
#Task1
image= Image . open ("/content/drive/MyDrive/AI and ML/Spiderman.JPG") . convert("L")
# Convert to NumPy array
image_array = np . array ( image )
# Get image dimensions
height , width = image_array . shape
# Reshape the image into a 2D array where each row represents a row of pixels
# This allows PCA to process each row as an observation and find patterns across pixel intensities
data = image_array . reshape ( height , width )
# Compute the mean of each column ( feature )
mean = np . mean ( data , axis =0)
# Subtract mean to center the data
centered_data = data - mean
# Compute covariance matrix
cov_matrix = np . cov ( centered_data , rowvar = False )
print("Covariance Matrix:")
print(cov_matrix)
print("Centered Data:")
print(centered_data)
```

```python
#Task2
# Compute eigenvalues and eigenvectors
eigenvalues , eigenvectors = np . linalg . eigh ( cov_matrix )
# Sort eigenvalues and corresponding eigenvectors in descending order
sorted_indices = np . argsort ( eigenvalues ) [:: -1]
eigenvalues = eigenvalues [ sorted_indices ]
eigenvectors = eigenvectors [: , sorted_indices ]
print("Eigenvalues:")
print(eigenvalues)
print("Eigen Vectors:")
print(eigenvectors)
# Plot the explained variance ratio
explained_variance_ratio = eigenvalues / np .sum ( eigenvalues )
plt . plot ( np . cumsum ( explained_variance_ratio ) )
plt . title (" Cumulative Explained Variance ")
plt . xlabel (" Number of Components ")
plt . ylabel (" Cumulative Explained Variance ")
plt . grid ( True )
plt . show ()
```

```python
#Task-3
k_values = [10, 50, 100, 500]
for k in k_values:
    # Select the top k eigenvectors as principal components
    components_k = eigenvectors[:, :k]

    # Project the centered_data onto these k principal components
    compressed_data_k = np.dot(centered_data, components_k)

    # Reconstruct the image from the compressed data
    decompressed_data_k = np.dot(compressed_data_k, components_k.T) + mean

    # Reshape the decompressed data back to image dimensions
    decompressed_image_k = decompressed_data_k.reshape(height, width)

    # Display the original, compressed, and reconstructed images for the current k
    plt.figure(figsize=(15, 6))

    plt.subplot(1, 3, 1)
    plt.imshow(image_array, cmap="gray")
    plt.title("Original Image")
    plt.axis("off")

    plt.subplot(1, 3, 2)
    # The compressed data is typically a lower-dimensional representation
    # We display it to show its existence, but it's not a direct 'image' in the visual sense
    # We might need to scale it for better visualization if desired, but for now just showing the structure
    plt.imshow(compressed_data_k, cmap="gray", aspect="auto")
    plt.title(f"Compressed Representation (k={k})")
    plt.axis("off")

    plt.subplot(1, 3, 3)
    plt.imshow(decompressed_image_k, cmap="gray")
    plt.title(f"Reconstructed Image (k={k})")
    plt.axis("off")

    plt.tight_layout()
    plt.show()
```
